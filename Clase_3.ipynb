Actividad Práctica 2: Optimización de Algoritmos

```python
def has_duplicate(arr):
    seen = set()
    for num in arr:
        if num in seen:
            return True
        seen.add(num)
    return False
```

Utilizamos un conjunto (set) para almacenar los elementos ya vistos, lo que permite optimizar el algoritmo a O(n).
